```python
X = [1,34,56,78,90]
y =[1,2,4,6,8]
```

```python
# get m
def get_m(X,y):
    return (np.corrcoef(X,y)*(np.std(y)/np.std(X)))
```

```python
def get_c(X,y):
    return  np.mean(y) - (get_m(X,y)*np.mean(X))
```

```python
class LinearReg(BaseModel):
    def __init__(self,):
        print("i was created")
        
    def get_m(self):
        m = (np.corrcoef(self.X,self.y)*(np.std(self.y)/np.std(self.X)))
        return m[0][1]
    
    def get_c(self):
        c = (np.mean(self.y) - (get_m(self.X,self.y)*np.mean(self.X)))
        return  c[0][1]
    
    def fit(self,X,y):
        self.X = X
        self.y = y
        
        c = self.get_c()
        m = self.get_m()
        return f"Model Trained c = {c} m = {m}"
    
    def predict(self, X):
        y_pred = []
        for x in X : 
           y_pred.append(self.get_m()*x + self.get_c())
           
        return y_pred
    
    def MAE(self, y_true, y_pred):
        y_true = np.array(y_true)
        y_pred = np.array(y_pred)
        
        MAE = np.mean(np.abs(y_true-y_pred))
        
        return f"MAE = {MAE}"
```

```python
model = LinearReg()
```

```python
model.fit(X,y)
```
